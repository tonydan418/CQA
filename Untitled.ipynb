```python
#check time and subset dataframe
#end_date_time in datetime.datetime(2017,3,28,0,0)
def extract_time(df,end_date_time):
    #convert datetime to datetime64
    dt = np.datetime64(end_date_time)
    #dt_1 = np.datetime64(start_date_time)
    i = 0
    while df.iloc[i,0] <= dt:
        i = i +1
        
    return df.iloc[0:i,:]

#clean client name
#def clean_name(df):
#    df.iloc[:,1]


#change buy/sell to 1/-1
#######################################
def long_short_convert(df):
    
    v = list(df.iloc[:,6])
    v_1 = v
    for i in range(0,len(v)):
        if v[i]== "BUY":
            v_1[i] = 1
        elif v[i] == "SELL":
            v_1[i] = -1
        else:
            v_1[i]=0
    df.iloc[:,6] = v
    return df
            
            
#calculate trade positions within each date
def trade_positions(df,end_date_time, trade_pos):
    #for j in range():
    dt = np.datetime64(end_date_time)
    df_1 = df[df.iloc[:,0] <= dt]
    
    for i in range(0,len(df_1.iloc[:,0])):
        #client name
        client_name = df_1.iloc[i,1]
        #bond name
        bond_name = df_1.iloc[i,4]
        #long/short variable
        l_s = df_1.iloc[i,6]
        #trading volume
        volume = df_1.iloc[i,10]
        
        ## add new client to the list
        if client_name not in trade_pos:
            trade_pos[client_name] = {}
            trade_pos[client_name][bond_name] = l_s*volume
        ##add new bond to existing client
        elif client_name in trade_pos and bond_name not in trade_pos[client_name]:
            trade_pos[client_name][bond_name] = l_s*volume
        ##adjust trade position to existing bond of existing client
        elif client_name in trade_pos and bond_name in trade_pos[client_name]:
            trade_pos[client_name][bond_name] = trade_pos[client_name][bond_name] + l_s*volume
        else:
            print("not valid")
        print(i)
        #print(len(df_1.iloc[:,0]))
        print(df_1.iloc[i,0])
        #i = i+1
        
    return trade_pos

```
